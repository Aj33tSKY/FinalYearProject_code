```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

accuracy_matrix = np.array([
    [97.00, 65.00], 				
    [89.50, 92.00], 						
])

# Dataset labels
dataset_labels = ["4B-2D-Uni", "10B-2D-Uni"]
#dataset_labels = ["4B-2D-Uni", "4B-3D-Uni", "4B-2D-Non", "4B-3D-Non"]

plt.figure(figsize=(8, 6))
sns.heatmap(accuracy_matrix, annot=True, fmt=".2f", cmap="Blues",
            xticklabels=dataset_labels, yticklabels=dataset_labels, linewidths=0.5, vmin=50, vmax=100)

plt.xlabel("Test Set")
plt.ylabel("Trained Model")
plt.title("Inter-Class Testing Accuracy Matrix")
plt.show()
```
